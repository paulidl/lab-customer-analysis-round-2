```python
#Drop the first column:
display(file.columns)
file = file.drop('Unnamed: 0', axis=1)
```

```python
#Fix the column that is not standard:
file.rename(columns={"EmploymentStatus": "employment_status"}, inplace=True)
display(file.columns)
```

```python
#Standarize the labels:
def standard_col(x):
    cols = []
    for item in x:
        col = ""
        for letter in item:
            if letter == " ":
                letter = "_"
                col += letter
            else:
                col += letter
        cols.append(col.lower())
    return cols

file.columns = standard_col(file.columns)

display(file.columns)
```

```python
## 6. Datetime format - Extract the months from the dataset and store in a separate column. 
#Then filter the data to show only the information for the first quarter , ie. January, February and March. 
#Hint: If data from March does not exist, consider only January and February.


file['Effective_To_Date'] = pd.to_datetime(file['Effective_To_Date'], errors='coerce')
file['Effective_To_Date']
```
